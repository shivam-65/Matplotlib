```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=3*x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x-lable")
plt.ylabel("y-lable")
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=3)
plt.show()
```

```python
z=4*x
z
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=3)
plt.plot(x,z,color='g',linestyle='-',linewidth=2)
plt.title("line plot")
plt.xlabel("x-lable")
plt.ylabel("y-lable")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.title("line plot")
plt.xlabel("x-lable")
plt.ylabel("y-lable")

plt.subplot(2,1,2)
plt.plot(x,z,color='r',linestyle=':',linewidth=2)
plt.title("line plot")
plt.xlabel("x-lable")
plt.ylabel("z-lable")

plt.show()
```

```python
student={"shivam":90,"akshay":80,"yash":70}

names=list(student.keys())
values=list(student.values())

plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color="r")
plt.title("Bar Plot")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```
